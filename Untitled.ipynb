```python
import pandas as pd
import yfinance as yf
import numpy as np

returns = pd.read_csv('returns.csv')
returns.columns
```

```python
import yfinance as yf
import pandas as pd
import time

tickers = list(returns.columns)[1:]
market_caps = {}

for ticker in tickers:
    try:
        info = yf.Ticker(ticker).info
        cap = info.get('marketCap', 0)  
        market_caps[ticker] = cap
          
    except Exception as e:
        print(f"Error retrieving {ticker}: {e}")
        market_caps[ticker] = 0


# Convert to a DataFrame 
market_caps_df = pd.DataFrame.from_dict(market_caps, orient='index', columns=['MarketCap'])
print(market_caps_df)

#Sort by market cap
market_caps_df = market_caps_df.sort_values(by="MarketCap", ascending=False)



# Number of groups you want
groups_cnt = 5
n_rows = len(market_caps_df)
group_length = n_rows // groups_cnt

# Create group labels
groups = np.repeat(np.arange(1, groups_cnt + 1), group_length)
remainder = n_rows - len(groups)
if remainder > 0:
    groups = np.append(groups, [groups_cnt] * remainder)

# Add groups column
market_caps_df["MarketCapGroup"] = groups

market_caps_df
```
